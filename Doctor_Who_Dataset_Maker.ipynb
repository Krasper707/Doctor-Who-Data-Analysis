```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import time
import re
```

```python
url="https://www.imdb.com/title/tt0436992/episodes/?season=1&ref_=ttep_ep_sn_pr"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
}
response = requests.get(url, headers=headers)

soup=BeautifulSoup(response.content,"html.parser")
```

```python
ans=[]
for i in range(1,14):
  url=f"https://www.imdb.com/title/tt0436992/episodes/?season={i}&ref_=ttep_ep_sn_pr"
  headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36"
  }
  response = requests.get(url, headers=headers)

  soup=BeautifulSoup(response.content,"html.parser")
  articles=soup.find_all("div",class_="sc-f2169d65-4 kBjDMi")
  titles=soup.find_all("div",class_="sc-f2169d65-5 kNicn")
  res=[]
  for j in range(len(articles)):

    te=titles[j].text.split("âˆ™")
    seep=te[0]
    se,ep=seep.split(".")
    air_date = articles[j].find("span", class_="sc-f2169d65-10 bYaARM").text
    rating=articles[j].find("span",class_="ipc-rating-star--rating").text
    totalvotes=articles[j].find("span",class_="ipc-rating-star--voteCount").text
    days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

    title = re.sub(rf"({'|'.join(days)})$", "", te[1]).strip()

    res.append([se,ep,title,air_date,rating,totalvotes])
    time.sleep(1)
  ans.append(res)
```

```python
e=ans.copy()
```

```python
ans=e.copy()
for i in range(len(ans)):
  for j in range(len(ans[i])):
    print(ans[i][j])
    days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]

    pattern = rf"({'|'.join(days)})"

    match = re.split(pattern, ans[i][j][2], maxsplit=1)
    title = match[0].strip()
    date = match[1].strip() + match[2].strip() if len(match) > 2 else ""  # "Fri, Mar 17, 2006"
    ans[i][j][2] = title
    ans[i][j][3] = date
```

```python
columns = ["Season", "Episode Number", "Episode Name", "Air Date", "IMDb Rating", "Votes"]
r = [j for i in ans for j in i]  # Flattening the nested list

df = pd.DataFrame(r, columns=columns)


if not df["Votes"].isnull().all():  # Check if column is not completely empty
    df["Votes"] = df["Votes"].astype(str).apply(lambda x: x.split()[-1] if x != 'None' else None)

df
```
